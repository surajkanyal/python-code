```python
# Sscraping the site I found
from bs4 import BeautifulSoup
import requests

response = requests.get('https://wynk.in/music/playlist/biggest-hits-of-2010s-hindi/bb_1576563267338')

print(response)
```

```python
res_file = response.text
soup = BeautifulSoup(res_file,'html.parser')
print(soup.title)
```

```python
print(soup.prettify())
#This implies we have a whole bunch of data gathered
```

```python
# lets take out the top 100 songs
def songs():
    song =soup.find_all(name = 'a' , class_ = 'jsx-7c093e1a359b5feb text-title font-normal line-clamp-1 md:line-clamp-2 text-sm hover:underline')
    song_20 = [s.getText() for s in song]
    return song_20
```

```python
songs_20 = songs()
print(len(songs_20))
print(songs_20)
```
