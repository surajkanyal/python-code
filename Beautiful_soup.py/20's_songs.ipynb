```python
# Sscraping the site I found
from bs4 import BeautifulSoup
import requests

response = requests.get('https://www.jiosaavn.com/featured/hindi-00s/tsJahdem34A_')

print(response)
```

```python
res_file = response.text
soup = BeautifulSoup(res_file,'html.parser')
print(soup.title)
```

```python
print(soup.prettify())
#This implies we have a whole bunch of data gathered
```

```python
# lets take out the top 100 songs
def songs():
    song =soup.find_all(name = 'a',class_ = 'u-color-js-gray')
    song_rand = [s.getText() for s in song]
    return song_rand
```

```python
Music = songs()
Music = Music[:40]
```
